````r
# Check if the input is a .JSON file for provBookR
json.check <- function(file.name){    
    # Check if it has a javascript file extension
    message("CHECKING FILE...")
    if(grepl("\\.json", file.name)){
        return(TRUE)
    }
    else if(grepl("\\.r", file.name)){
        return(FALSE)
    }else{
        message("This is neither a .JSON or .R file")
    }
        
}

# Parse the provenance for provBookR
provBookR.parse <- function(file.name){
    # Extract important information from .json
    json.list <- prov.parse(file.name)

    # Filter for operations that are assignment
    proc.nodes <- get.proc.nodes(json.list)
    proc.nodes.assignment <- proc.nodes[grep("<-",proc.nodes$name), c("name","startLine","endLine","elapsedTime")]
    colnames(proc.nodes.assignment)[1] <- "expr"

    # Filter for data values : this will collect all of the data types (if it's data then it's been assignmed)
    data.nodes <- get.data.nodes(json.list)
    data.nodes.attr <- data.nodes[,c("name","value", "valType", "type")]
    data.nodes.attr <- data.nodes.attr[!(data.nodes.attr$type=="File"),]

    # Collect all the matrices
    ddg.sub.matrix <- cbind(proc.nodes.assignment,data.nodes.attr)
    ddg.sub.matrix <- ddg.sub.matrix[,c("name","expr","startLine","endLine","elapsedTime","value","valType","type")]
    return(ddg.sub.matrix)

}

# Generate the lineage for provBookR
provBookR.lineage <- function(file.name, node.id){   
    
    # Create adj.graph
    adj.graph <- create.graph(file.name, isFile = T)
    
    # Collect the lineage from the provenance
    lineage <- get.lineage(adj.graph, node.id, forward = F)
    return(lineage)
}

# Cleaner provBookR function
provBookR.prov.parse <- function(file.name, data.object.name){
    # Collect the parsed provenance
    parsed.prov <- get.variable.named(prov.run(file.name), data.object.name)
}
````

````r
# Generate page structure for provbook
page.maker <- function(x){
    # bookblock dependency
    dep.js.bookblock <- htmltools::htmlDependency("bookblock", "2.0.1", c(href="shared/js/bookblock.js"),
    script = "bookblock.js")

    # bookblock dependency (minified version)
    dep.js.bookblock.min <- htmltools::htmlDependency("bookblock", "2.0.1", c(href="shared/js/bookblock.min.js"),
    script = "bookblock.min.js")

    # JQuery dependency (minified version)
    dep.js.jquery.bookblock.min <- htmltools::htmlDependency("jqueryui", "2.0.1", c(href="shared/js/jquery.bookblock.min.js"),
    script = "jquery.bookblock.min.js")

    # JQuery-pp dependency
    dep.js.jquerypp.custom <- htmltools::htmlDependency("jqueryui", "1.0.0", c(href="shared/js/jquerypp.custom.js"),
    script = "jquerypp.custom.js")

    # Modernizr dependency
    dep.js.modernizr.custom.js <- htmltools::htmlDependency("jqueryui", "1.11.4", c(href="shared/js/modernizr.custom.js"),
    script = "modernizr.custom.js")

    # Load css dependencies
    dep.css.bookblock <- htmlDependency("bookblock", "1.0.0", c(href="shared/css/bookblock.css"), stylesheet ="bookblock.css")
    dep.css.default <- htmlDependency("default", "1.0.0", c(href="shared/css/default.css"), stylesheet = "default.css")
    dep.css.demo4 <- htmlDependency("demo4", "1.0.0", c(href="shared/css/demo4.css"), stylesheet = "demo4.css")

    # Store the dependencies in a list
    dependencies <- list(dep.js.bookblock, dep.js.bookblock.min, dep.js.jquery.bookblock.min, dep.js.jquerypp.custom,  
                           dep.js.modernizr.custom.js, dep.css.bookblock, dep.css.default, dep.css.demo4)
    
    # Generate html structure for pages
    provBookR.pages <- tags$div(class ="bb-item",
                               div(class ="bb-custom-side",
                                  p(DT::datatable(x, options = list(dom = 't'), filter = list(position = "top")))),
                               div(class ="bb-custom-side",
                                  p(DT::datatable(x, options = list(dom = 't'), filter = list(position = "top")))),
                               )
    # Attach the dependencies to the provbook
    provBookR.pages <- htmltools::attachDependencies(provBookR.pages,dependencies)   
    provBookR.pages
}
````

````r
prov.browser <- function(file.name){
    
    # Start browser loop...
    UI.condition <- TRUE
    while(UI.condition){
        
        # Complete browser initializing
        initial.options.message <- "provBookR browser initialized, type \"help\" for more information or Q to quit"
        initial.prompt <- paste(initial.options.message, script.name,sep="\n")
        message(initial.prompt)
        
        # Collecting initial user input
        UI.condition <- tolower(readline())
        
        # Parse the users input
        switch(UI.condition
               "Q" = UI.user.input.quit(),
               "help" = UI.user.input.help(),
               "TF" = UI.user.input.TF(),
               "TB" = UI.user.input.TB(),
               "BG" = UI.user.input.BG(),
               message(initial.prompt)
      )
    }
}

# browser "quit" helper function                                
UI.user.input.quit  <- function(){
    message("Are you sure you want to quit (Y/N):")
    user.input <- tolower(readline())
    if(user.input == "y"){
            return(!UI.condition)
    }
}
                                
# browser "help" helper function
UI.user.input.help  <- function(){
    cat("provBookR operations:\n")
    options <- c("Quit provBookR", "provBookR operations", "Collect forward provenance",
    "Collect backward provenance", "Generate provbook for script")
    options.df <- data.frame(words = words, sep = ':', nchar = c("\"Q\"", "\"help\"", "\"TF\"", "\"TB\"","\"BG\""))
    co <- capture.output(gdata::write.fwf(options.df, width = c(max(nchar(options)), 1, 6), justify = 'l'))[-1]
    cat(co, sep = '\n')
}

# browser "TF" helper function                                
UI.user.input.TF  <- function(){
    message(paste("Collecting forward provenance for:", data.object.name, sep=" ")
    print(provBookR.lineage(file.name, node.id, "T")) # not sure where I get node.id from...
    message(paste("Would you like to generate a provbook for the forward provenance (Y/N)", sep=" "))            
    user.input <- tolower(readline())
    
    # Handle user input
    if(user.input == "y"){
        message("Generating provbook...")
        provBookR.prov.visualization()
    }       
}

# browser "TB" helper function            
UI.user.input.TB  <- function(){
    message(paste("Collecting backward provenance for:", data.object.name, sep=" ")
    print(provBookR.lineage(file.name, node.id, "B")) # not sure where I get node.id from...
    message(paste("Would you like to generate a provbook for the forward provenance (Y/N)", sep=" "))            
    user.input <- tolower(readline())
    
    # Handle user input
    if(user.input == "y"){
        message("Generating provbook...")
        provBookR.prov.visualization()
    }       

}

# browser "BG" helper function
UI.user.input.BG  <- function(){
    message(paste("Generating provbook for:", data.object.name, sep=" "))
    print(provBookR.lineage(file.name, node.id, "B"))

}
            
# browser "help" helper function
UI.user.input.help  <- function(){
    cat("provBookR operations:\n")
    options <- c("Quit provBookR", "provBookR operations", "Collect forward provenance",
    "Collect backward provenance", "Generate provbook for script")
    options.df <- data.frame(words = words, sep = ':', nchar = c("\"Q\"", "\"help\"", "\"TF\"", "\"TB\"","\"BG\""))
    co <- capture.output(gdata::write.fwf(options.df, width = c(max(nchar(options)), 1, 6), justify = 'l'))[-1]
    cat(co, sep = '\n')
}
            
script.name <- "test.script.r"
initial.options.message <- "provBookR browser initialized, type \"help\" for more information or \"Q\" to quit"
initial.prompt <- paste(initial.options.message, script.name,sep="\n")
message(initial.prompt)
````
